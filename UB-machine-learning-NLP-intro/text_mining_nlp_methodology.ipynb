## 1.1 Remove accents
Sometimes it is useful to remove accents and other marks from our text

````
import string
import unicodedata
def remove_accents(text):
    '''
    Tokenize a string in Spanish
    Parameters
    ----------
    text : str
        String I want to process
    Returns
    -------
    String without accents
    '''
    return ''.join(x for x in unicodedata.normalize('NFKD', text) if x in string.printable)

print(f"I am removing characters not present in this list: {string.printable}")
````

````
print(f"WITH accents: \n{text}")
print(f"WITHOUT accents: \n{remove_accents(text)}")
````

## 1.2 Tokenize
Token: sequence of characters in some particular document that are grouped together as a useful semantic unit for processing

````
# I need to download some packages to use the tokenization function
!pip install spacy==2.3.5
!python -m spacy download es_core_news_sm
!python -m spacy download es_core_news_lg
````

````
import es_core_news_sm
nlp = es_core_news_sm.load()
def tokenize(text):
    '''
    Tokenize a string
    Parameters
    ----------
    text : str
        Text string to tokenize.
    Returns
    -------
    tokenized : list
        List of tokens (includes punctuation tokens).
    '''
    doc = nlp(text)
    token_list = []
    for token in doc:
        token_list.append(token.text)
    return token_list
````

````
print(f"Original string: \n{text}")
print(f"Tokens: \n{tokenize(text)}")
````

## 1.3 Lowercase
It is also useful to lowercase all words (there is no difference in meaning between "Cat" and "cat)". However, casing is relevant for recognizing acronyms. Then, I usually only lowercase words with less than 4 letters.

````
def lowercase(list_of_tokens, min_upper=4):
    '''
    Clean sting: lowercase, remove extra whitespaces, remove punctuation and remove accents.
    
    Parameters
    ----------
    list_of_tokens: list
    min_upper: int. 
        It specifies the minimum number of characters of a word to lowercase 
        (to prevent mistakes with acronyms).
    
    Returns
    -------
    list of lowercased tokens
    '''
    return list(map(lambda x: x.lower() if len(x)>min_upper else x, list_of_tokens))
````

````
print(f"Original tokens: \n{tokenize(text)}")
print(f"Lowercased tokens: \n{lowercase(tokenize(text), 4)}")
````

### 3.1.2 Transform Gold Standard format into Spacy format
The annotation files that we saw (.ann) are not properly formatted for the Python library we are going to use to train our model. Then, I give you some functions to transform between the two formats (no need at all to understand these functions!)

````
import time
import os
import pandas as pd
from nltk.tokenize.punkt import PunktSentenceTokenizer

def parse_ann(datapath, labels_to_ignore = [], with_notes=False):
    '''
    DESCRIPTION: parse information in .ann files.
    
    Parameters
    ----------
    datapath: str. 
        Route to the folder where the files are. 
           
    Returns
    -------
    df: pandas DataFrame 
        It has information from ann files. Columns: 'annotator', 'bunch',
        'filename', 'mark', 'label', 'offset1', 'offset2', 'span', 'code'
    '''
    start = time.time()
    info = []
    ## Iterate over the files and parse them
    filenames = []
    print(datapath)
    for root, dirs, files in os.walk(datapath):
        for filename in files:
            if filename[-3:] != 'ann':
                continue
            info = parse_one_ann(info, filenames, root, filename, labels_to_ignore,
                                  ignore_related=True, with_notes=with_notes)
             
    if with_notes == True:
        df = pd.DataFrame(info, columns=['annotator','bunch','filename','mark',
                                     'label','offset1','offset2','span','code'])
    else:
        df = pd.DataFrame(info, columns=['annotator','bunch','filename','mark',
                                     'label','offset1','offset2','span'])
    end = time.time()
    print("Elapsed time in parsing Brat files: " + str(round(end-start, 2)) + 's')
    df[["offset1", "offset2"]] = df[["offset1", "offset2"]].apply(pd.to_numeric)

    return df

def parse_one_ann(info, filenames, root, filename, labels_to_ignore,
                  ignore_related=True, with_notes=False):
    
    f = open(os.path.join(root,filename)).readlines()
    filenames.append(filename)
    # Get annotator and bunch
    bunch = root.split('/')[-1]
    annotator = root.split('/')[-2][-1]

    # Parse .ann file
    if with_notes==True:
        mark2code = {}
        for line in f:
            if line[0] != '#':
                continue
            line_split = line.split('\t')
            mark2code[line_split[1].split(' ')[1]] = line_split[2].strip()
    
    for line in f:
        if line[0] != 'T':
            continue
        splitted = line.split('\t')
        if len(splitted)<3:
            print('Line with less than 3 tabular splits:')
            print(root + filename)
            print(line)
            print(splitted)
        if len(splitted)>3:
            print('Line with more than 3 tabular splits:')
            print(root + filename)
            print(line)
            print(splitted)
        mark = splitted[0]
        label_offset = splitted[1].split(' ')
        label = label_offset[0]
        # Only store labels I am interested in
        if label in labels_to_ignore:
            continue
        offset = label_offset[1:]
        span = splitted[2].strip('\n')
        if with_notes==False:
            info.append([annotator, bunch, filename,mark, label,
                         offset[0], offset[-1], 
                         span.strip('\n')])
            continue
        
        if mark in mark2code.keys():
            code = mark2code[mark]
        else:
            code = ''
        info.append([annotator, bunch, filename,mark, label,
                     offset[0], offset[-1], 
                     span.strip('\n'), code])
            
    return info

def brat2spacy(datapath, labels_to_ignore = [], with_notes=False):
    
    # Parse Brat data
    ann = parse_ann(datapath, labels_to_ignore = [], with_notes=False)
    ann = ann[['filename', 'label', 'offset1', 'offset2', 'span']].copy()
    
    # Brat to Spacy
    spacy_format = []
    filenames = os.listdir(datapath)
    for filename in filenames:
        if filename.endswith('.ann'):
            continue
        ann_filename = os.path.splitext(filename)[0] + '.ann' # Find ANN filename
    
        txt = open(os.path.join(datapath, filename)).read() # Read TXT
        
        for start, end in PunktSentenceTokenizer().span_tokenize(txt): # Split sentences
            sentence = txt[start:end]
            sentence_processed = ' '.join(lowercase(tokenize(remove_accents(sentence)))) # Do the same preprocessing as before!
            entities = []
            for idx,row in (ann.loc[(ann['offset1']>=start) &
                                    (ann['offset1']<=end) & 
                                    (ann['filename']==ann_filename),:]).iterrows():
                # Store entities of this sentence
                span_processed = ' '.join(lowercase(tokenize(remove_accents(row['span']))))
                offset1 = sentence_processed.find(span_processed)
                offset2 = offset1 + len(span_processed)
                entities.append((offset1, offset2, row['label']))
            if len(entities) == 0:
                continue

            entities = delete_overlapped_annotations(entities)
            spacy_format.append((sentence_processed, {"entities": entities}))
    return spacy_format

def delete_overlapped_annotations(annotations):
    df = pd.DataFrame(annotations, columns = ['offset1', 'offset2', 'label'])
    to_delete = []
    for idx,row in df.iterrows():
        other_rows = df.loc[df.index!=idx,:]
        if any(other_rows.apply(lambda x: (x['offset1']<=row['offset1']) &
                                (x['offset2']>=row['offset2']), axis=1)):
            to_delete.append(idx)
        if any(other_rows.apply(lambda x: (x['offset1']>=row['offset1']) &
                                (x['offset2']>=row['offset2']), axis=1)):
            to_delete.append(idx)
        
    final_df = df.drop(to_delete).copy()
    
    return list(map(lambda x: tuple(x), final_df.values.tolist()))
````

## 3.2 Train Machine Learning algorithm
inspired by https://www.machinelearningplus.com/nlp/training-custom-ner-model-in-spacy/
and https://towardsdatascience.com/custom-named-entity-recognition-using-spacy-7140ebbb3718

Load machine learning algorithm. We will train it with the Gold Standard that we downloaded in the previous section

````
import es_core_news_lg
nlp = es_core_news_lg.load() 
print(nlp)
````

Configure the Python object. Basically, I tell the algorithm that we want to do Named Entity Recognition and which are the labels we are using here (either MORFOLOGIA_NEOPLASIA, which means an entity related with cancer morphology, or Null [O]). (No need at all to understand these 2 blocks of code!)

````
if 'ner' not in nlp.pipe_names:
    ner = nlp.create_pipe('ner')
    nlp.add_pipe(ner)
else:
    ner = nlp.get_pipe('ner')
print(nlp.pipe_names)
````

````
# Add new entity labels to entity recognizer
LABEL = ['MORFOLOGIA_NEOPLASIA', 'O']
for i in LABEL:
    ner.add_label(i)
# Inititalizing optimizer
optimizer = nlp.begin_training()
````

And now, let's train the machine learning model. We input the Gold Standard 10 times (```n_iter=10```) and the algorithm starts finding patterns in the annotated Gold Standard



````
import random
from spacy.util import minibatch, compounding
# Get names of other pipes to disable them during training to train # only NER and update the weights
other_pipes = [pipe for pipe in nlp.pipe_names if pipe != 'ner']
n_iter=10
with nlp.disable_pipes(*other_pipes):  # only train NER
  for itn in range(n_iter):
    random.shuffle(TRAIN_DATA)
    losses = {}
    batches = minibatch(TRAIN_DATA, size=compounding(4., 32., 1.001))
    for batch in batches:
      texts, annotations = zip(*batch)
      # Updating the weights
      nlp.update(texts, annotations, sgd=optimizer, drop=0.35, losses=losses)
    print('Losses', losses)
````

Text representation is integrated in Spacy environment! Then, in this case, we do not need to transform the tokens into vectors, Spacy does it internally!
